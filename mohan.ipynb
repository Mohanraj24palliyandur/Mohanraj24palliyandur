```python
df2=pd.read_csv('/content/historical_data.csv')
```

```python
df2.head()
```

```python
df2.info()
```

```python
df2.isnull().sum()
```

```python
df2['time'] = pd.to_datetime(df2['Timestamp IST'], format='%d-%m-%Y %H:%M')
df2['date'] = df2['time'].dt.date
df2['date'] = pd.to_datetime(df2['date'])
```

```python
df2['Account'].nunique()
df2[['Size USD', 'Execution Price']].describe()
df2['Side'].value_counts()
plt.figure(figsize=(10,5))
sns.histplot(df2['Closed PnL'], bins=50, kde=True)
plt.title('Distribution of Closed PnL')
plt.savefig('/content/closed_pnl_distribution.png')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.histplot(df2['Closed PnL'], bins=100, kde=True)
plt.xlim(-500, 500)  # Zoom to focus on common values
plt.title('Zoomed Distribution of Closed PnL (-500 to 500)')
plt.savefig('/content/closed_pnl_distribution_zoomed.png')
plt.show()
```

```python
# Remove extreme outliers using quantiles
filtered_df = df2[(df2['Closed PnL'] > df2['Closed PnL'].quantile(0.05)) &
                  (df2['Closed PnL'] < df2['Closed PnL'].quantile(0.95))]

plt.figure(figsize=(10,5))
sns.histplot(filtered_df['Closed PnL'], bins=50, kde=True)
plt.title('Distribution of Closed PnL (Without Outliers)')
plt.savefig('/content/closed_pnl_distribution_no_outliers.png')
plt.show()
```

```python


# Avoid log(0) or negative by adding 1 and taking absolute
df2['log_pnl'] = np.sign(df2['Closed PnL']) * np.log1p(np.abs(df2['Closed PnL']))

plt.figure(figsize=(10,5))
sns.histplot(df2['log_pnl'], bins=50, kde=True)
plt.title('Log-Transformed Closed PnL Distribution')
plt.savefig('/content/log_closed_pnl.png')
plt.show()
```
